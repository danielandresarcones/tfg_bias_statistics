# Import Required Libraries
Import the necessary libraries, including NumPy and Matplotlib.

```python
# Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
from ishigami_function_3d import IshigamiFunction, IshigamiFunctionPolynomial
```

# Instantiate Ishigami Function
Create an instance of the IshigamiFunction class with default or custom parameters.

```python
# Instantiate Ishigami Function

# Create an instance of the IshigamiFunction class with default parameters
ishigami_default = IshigamiFunction()

# Create an instance of the IshigamiFunction class with custom parameters
ishigami_custom = IshigamiFunction(a=5, b=0.2)

# Display the instances to verify
ishigami_default, ishigami_custom
```

# Evaluate Ishigami Function
Evaluate the Ishigami function at various 3-dimensional input vectors and display the results.

```python
# Evaluate Ishigami Function

# Define a list of 3-dimensional input vectors
input_vectors = [
    [np.pi / 2, np.pi / 2, np.pi / 2],
    [np.pi / 4, np.pi / 4, np.pi / 4],
    [np.pi, np.pi, np.pi],
    [0, 0, 0],
    [1, 2, 3]
]

# Evaluate the Ishigami function at each input vector using the default parameters
results_default = [ishigami_default.evaluate(np.array(x).reshape((1,-1))) for x in input_vectors]

# Evaluate the Ishigami function at each input vector using the custom parameters
results_custom = [ishigami_custom.evaluate(np.array(x).reshape((1,-1))) for x in input_vectors]

# Display the results
for i, x in enumerate(input_vectors):
    print(f"Input vector: {x}")
    print(f"Default parameters result: {results_default[i]}")
    print(f"Custom parameters result: {results_custom[i]}")
    print()
```

# Visualize Ishigami Function
Use Matplotlib to visualize the Ishigami function over a range of input values.

```python
# Visualize Ishigami Function

# Define the range of input values for x1, x2, and x3
x1 = np.linspace(-np.pi, np.pi, 100)
x2 = np.linspace(-np.pi, np.pi, 100)
x3 = np.linspace(-np.pi, np.pi, 100)

# Create a meshgrid for x1, x2, and x3
X1, X2, X3 = np.meshgrid(x1, x2, x3, indexing="ij")

# Reshape X1, X2, and X3 to a single array of shape (100*100*100, 3)
input_array = np.array([X1.flatten(), X2.flatten(), X3.flatten()]).T
# Evaluate the Ishigami function over the meshgrid using the default parameters
Z_default = ishigami_default.evaluate(input_array).reshape(X1.shape)

# Plot the Ishigami function using the default parameters
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1[:, :, 20], X2[:, :, 20], Z_default[:, :, 20], cmap='viridis')

# Set plot labels and title
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Ishigami Function Value')
ax.set_title('Ishigami Function Visualization (Default Parameters)')

# Show the plot
plt.show()
```

```python
# Plot the Ishigami function using the default parameters for (x1, x3)
fig1 = plt.figure(figsize=(10, 7))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(X1[:, 0, :], X3[:, 0, :], Z_default[:, 0, :], cmap='viridis')

# Set plot labels and title
ax1.set_xlabel('x1')
ax1.set_ylabel('x3')
ax1.set_zlabel('Ishigami Function Value')
ax1.set_title('Ishigami Function Visualization (x1, x3)')

# Show the plot
plt.show()

# Plot the Ishigami function using the default parameters for (x2, x3)
fig2 = plt.figure(figsize=(10, 7))
ax2 = fig2.add_subplot(111, projection='3d')
ax2.plot_surface(X2[0, :, :], X3[0, :, :], Z_default[0, :, :], cmap='viridis')

# Set plot labels and title
ax2.set_xlabel('x2')
ax2.set_ylabel('x3')
ax2.set_zlabel('Ishigami Function Value')
ax2.set_title('Ishigami Function Visualization (x2, x3)')

# Show the plot
plt.show()
```

```python
# Generate predictions for different (a, b) parameters
ishigami_params = [
    IshigamiFunction(a=7, b=0.1),
    IshigamiFunction(a=5, b=0.1),
    IshigamiFunction(a=7, b=0.05)
]

# Evaluate the Ishigami function over the meshgrid for each parameter set
input_array = np.array([X1.flatten(), X2.flatten(), X3.flatten()]).T
Z_params = [ishigami.evaluate(input_array).reshape(X1.shape) for ishigami in ishigami_params]
```

```python

# Plot the Ishigami function for each parameter set
fig, axes = plt.subplots(3, 3, figsize=(15, 15), subplot_kw={'projection': '3d'})

# Titles for the subplots
titles = [
    'Ishigami Function (a=7, b=0.1)',
    'Ishigami Function (a=5, b=0.1)',
    'Ishigami Function (a=7, b=0.05)'
]

# Plot for (x1, x2)
for i in range(3):
    axes[i, 0].plot_surface(X1[:, :, 20], X2[:, :, 20], Z_params[i][:, :, 20], cmap='viridis', vmin=-2, vmax=8)
    axes[i, 0].set_xlabel('x1')
    axes[i, 0].set_ylabel('x2')
    axes[i, 0].set_zlabel('Función Ishigami')
    # axes[i, 0].set_zlabel('Ishigami Function Value')
    axes[i, 0].set_title(titles[i] + ' (x1, x2)')
    axes[i, 0].set_zlim(-2, 8)

# Plot for (x1, x3)
for i in range(3):
    axes[i, 1].plot_surface(X1[:, 0, :], X3[:, 0, :], Z_params[i][:, 0, :], cmap='viridis', vmin=-10, vmax=10)
    axes[i, 1].set_xlabel('x1')
    axes[i, 1].set_ylabel('x3')
    # axes[i, 1].set_zlabel('Ishigami Function Value')
    axes[i, 1].set_zlabel('Función Ishigami')
    axes[i, 1].set_title(titles[i] + ' (x1, x3)')
    axes[i, 1].set_zlim(-10, 10)

# Plot for (x2, x3)
for i in range(3):
    axes[i, 2].plot_surface(X2[0, :, :], X3[0, :, :], Z_params[i][0, :, :], cmap='viridis', vmin=0, vmax=7)
    axes[i, 2].set_xlabel('x2')
    axes[i, 2].set_ylabel('x3')
    # axes[i, 2].set_zlabel('Ishigami Function Value')
    axes[i, 2].set_zlabel('Función Ishigami')
    axes[i, 2].set_title(titles[i] + ' (x2, x3)')
    axes[i, 2].set_zlim(0, 7)

# Adjust layout
plt.tight_layout()
plt.savefig('./output/figures/ishigami_function_visualization.pdf')
plt.savefig('./output/figures/ishigami_function_visualization.pdf')
```

```python
# Define the range of c values
c_range = range(5)

# Generate predictions for different (a, b, c) parameters using IshigamiFunctionPolynomial
ishigami_poly_params = [
    IshigamiFunctionPolynomial(a=7, b=0.1, c=c) for c in c_range
]

# Evaluate the Ishigami function polynomial over the meshgrid for each parameter set
input_array = np.array([X1.flatten(), X2.flatten(), X3.flatten()])
Z_poly_params = [ishigami_poly.evaluate(input_array).reshape(X1.shape) for ishigami_poly in ishigami_poly_params]

# Plot the Ishigami function polynomial for each parameter set
fig, axes = plt.subplots(len(c_range), 3, figsize=(15, 15), subplot_kw={'projection': '3d'})

# Titles for the subplots
titles_poly = [
    # f'Ishigami Function (a=7, b=0.1, degree={c})' for c in c_range
    f'Función Ishigami (a=7, b=0.1, grado={c})' for c in c_range
]

# Plot for (x1, x2)
for i in range(len(c_range)):
    axes[i, 0].plot_surface(X1[:, :, 20], X2[:, :, 20], Z_poly_params[i][:, :, 20], cmap='viridis', vmin=-2, vmax=8)
    axes[i, 0].set_xlabel('x1')
    axes[i, 0].set_ylabel('x2')
    # axes[i, 0].set_zlabel('Ishigami Function Value')
    axes[i, 0].set_zlabel('Función Ishigami')
    axes[i, 0].set_title(titles_poly[i] + ' (x1, x2)')
    axes[i, 0].set_zlim(-2, 8)

# Plot for (x1, x3)
for i in range(len(c_range)):
    axes[i, 1].plot_surface(X1[:, 0, :], X3[:, 0, :], Z_poly_params[i][:, 0, :], cmap='viridis', vmin=-10, vmax=10)
    axes[i, 1].set_xlabel('x1')
    axes[i, 1].set_ylabel('x3')
    # axes[i, 1].set_zlabel('Ishigami Function Value')
    axes[i, 1].set_zlabel('Función Ishigami')
    axes[i, 1].set_title(titles_poly[i] + ' (x1, x3)')
    axes[i, 1].set_zlim(-10, 10)

# Plot for (x2, x3)
for i in range(len(c_range)):
    axes[i, 2].plot_surface(X2[0, :, :], X3[0, :, :], Z_poly_params[i][0, :, :], cmap='viridis', vmin=0, vmax=7)
    axes[i, 2].set_xlabel('x2')
    axes[i, 2].set_ylabel('x3')
    # axes[i, 2].set_zlabel('Ishigami Function Value')
    axes[i, 2].set_zlabel('Función Ishigami')
    axes[i, 2].set_title(titles_poly[i] + ' (x2, x3)')
    axes[i, 2].set_zlim(0, 7)

# Adjust layout
plt.tight_layout()
plt.savefig('./output/figures/ishigami_function_polynomial_visualization.pdf')
plt.savefig('./output/figures/ishigami_function_polynomial_visualization.pdf')
plt.show()
```
